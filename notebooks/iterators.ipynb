```python
# https://www.bilibili.com/video/BV1LzG1zqE7N/
# 有用的迭代器

from itertools import batched
numbers : list[int] = [1, 2, 3, 4, 5, 6, 7, 8, 9]
iter_batched = batched(numbers, n=4)
print(next(iter_batched))
print(next(iter_batched))
print(next(iter_batched))
print(next(iter_batched)) # iteration stop
```

```python
# 笛卡尔积
from itertools import product
colors: list[str] = ['r', 'g', 'b']
iter_product = product(colors, colors, colors)
for i in iter_product:
    print(''.join(i))
```

```python
# pytorch数据集加载
from torch.utils.data import DataLoader, Dataset, TensorDataset
import torch

num_samples = 1000
num_features = 6
X = torch.randn(num_samples, num_features)
y = torch.randint(0, 2, (num_samples,))  # 二分类

dataset = TensorDataset(X, y) # 可迭代对象
iter_samples = iter(dataset) # 返回一个迭代器对象
print(next(iter_samples))
print(next(iter_samples))
print(next(iter_samples))
print(next(iter_samples))

dataloader = DataLoader(dataset, batch_size=32, shuffle=True, num_workers=4) # 可迭代对象
iter_batches = iter(dataloader) # 返回一个迭代器对象
print(next(iter_batches))
print(next(iter_batches))
print(next(iter_batches))
```

```python
"""
Python 的 for 循环在底层会自动调用可迭代对象的 __iter__ 方法来获取一个迭代器，
然后通过 next() 方法逐一获取元素，直到抛出 StopIteration 异常。
"""

# 使用 for 循环迭代可迭代对象
for batch_data, batch_labels in dataloader:
    print(batch_data.shape, batch_labels.shape)
    break

# 使用 for 循环迭代迭代器
for batch_data, batch_labels in iter_batches:
    print(batch_data.shape, batch_labels.shape)
    break
```
